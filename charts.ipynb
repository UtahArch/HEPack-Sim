## Data Read and Prep

```python
data_mobile = {}

for filename in glob.glob("data_mobile/*/*"):
    with open(filename) as fin:
        temp = filename.split("/")
        model = tuple(temp[-2].split("_"))
        layer = temp[-1].split(".")[0]
#         print(model, layer)
        
        if model not in data_mobile:
            data_mobile[model] = {}
        data_mobile[model][layer] = {}

        lines = fin.readlines()
        i = 0

        layer_temp = layer.split("_")[:-6]
        layer_temp[0] = layer_temp[0][:-1] + " " + layer_temp[0][-1]
        layer_temp = " ".join(layer_temp)
        data_mobile[model][layer]['Layer'] = layer_temp

        data_mobile[model][layer]['Master Stats'] = []
        while "=" not in lines[i]:
            data_mobile[model][layer]['Master Stats'].append(lines[i].strip())
            i += 1

        h = -1
        for line in lines[i:]:
            line = line.strip()
            if "=" in line:
                h = line.split()[1]
                data_mobile[model][layer][h] = {}
            else:
                if "Data Movement" in line:
                    data_mobile[model][layer][h]["Data Movement"] = ast.literal_eval(line[(len("Data Movement      :	")):])
                else:
                    temp = [x.strip() for x in line.split(":")]
                    data_mobile[model][layer][h][temp[0]] = int(temp[1])
```

```python
data_resnet = {}

for filename in glob.glob("data_resnet/*/*"):
#     if 'FC' in filename:
#         continue
    with open(filename) as fin:
        temp = filename.split("/")
        model = tuple(temp[-2].split("_"))
        layer = temp[-1].split(".")[0]
#         print(model, layer)
        
        
        if model not in data_resnet:
            data_resnet[model] = {}
        data_resnet[model][layer] = {}
        
        lines = fin.readlines()
        i = 0

        layer_temp = layer.split("_")[:-6]
        layer_temp[0] = layer_temp[0][:-1] + " " + layer_temp[0][-1]
        layer_temp = " ".join(layer_temp)
        data_resnet[model][layer]['Layer'] = layer_temp

        data_resnet[model][layer]['Master Stats'] = []
        while "=" not in lines[i]:
            data_resnet[model][layer]['Master Stats'].append(lines[i].strip())
            i += 1
        
#         if layer == "CONV2_1_1_56_56_1_1_64_64":
#             print(filename)
        
        h = -1
        for line in lines[i:]:
            line = line.strip()
            if "=" in line:
                h = line.split()[1]
                data_resnet[model][layer][h] = {}
            else:                
                if "Data Movement" in line:
                    data_resnet[model][layer][h]["Data Movement"] = ast.literal_eval(line[(len("Data Movement      :	")):])
                else:
                    temp = [x.strip() for x in line.split(":")]
                    data_resnet[model][layer][h][temp[0]] = int(temp[1])
```
